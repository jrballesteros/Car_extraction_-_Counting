```python
# TFMS (4 a la vez) de la imagen, falta el random y el cambio en la iluminacion

mypath = r'C:\Users\john\Downloads\rocks - copia\coquina'
imagenes = 1

for f in os.listdir(mypath):
    i = Image.open(os.path.join(mypath, f))
    print(f, i.size, i.mode)
    fname, fext = os.path.splitext(f)
    i10 = i.rotate(10)
    i10.save(os.path.join(mypath,'rot10{}.jpg'.format(fname)))
    iflip = i.transpose(Image.FLIP_LEFT_RIGHT)
    iflip.save(os.path.join(mypath,'fhor{}.jpg'.format(fname)))
    ifvert = i.transpose(Image.FLIP_TOP_BOTTOM)
    ifvert.save(os.path.join(mypath,'fvert{}.jpg'.format(fname)))
    
    width, height = i.size
    m = 0.2
    xshift = abs(m) * width
    new_width = width + int(round(xshift))
    iwarp = i.transform((new_width, height), (Image.AFFINE),(1, m, -xshift if m > 0 else 0, 0, 1, 0),
                  (Image.BICUBIC))
    iwarp.save(os.path.join(mypath,'warp{}.jpg'.format(fname)))
    imagenes = imagenes + 1

print("El numero de imagenes procesadas fue de: "+ str((imagenes-1)*4))
```
