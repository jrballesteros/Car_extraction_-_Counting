```python
# Este script parte y nombra una ortofoto en tiles usando python y gdal retile

import os
from osgeo import gdal

rasterin_folder = 'C:/Users/john/Downloads/AutoMap/datos/area'
input_filename = 'RetiroAoI.tif'

out_path = 'C:/Users/john/Downloads/AutoMap/datos/tiles_600'


dset = gdal.Open(rasterin_folder + "/" + input_filename)

#print("La ortofoto de entrada es", dset.GetGeoTransform(), dset.GetProjection())

width = dset.RasterXSize
height = dset.RasterYSize

tile_width = 1000
tile_height = 1000

#print(width, 'x', height, ', tilesize:', tile_width)

for rasterin in os.listdir(rasterin_folder):
    if rasterin.endswith('.tif'):
        gdaltranString = 'gdal_retile.py -v -r bilinear -ps ' + str(tile_width) + ' ' \
        + str(tile_height) + ' -co ' + 'tiled=YES' + ' -targetDir ' + out_path + ' ' \
        + rasterin_folder + '/' + rasterin
        print(gdaltranString)
        os.system(gdaltranString)

print('El numero de imagenes producidas completas es:', (int(width/tile_width))**2)
```

```python
# Este script parte y nombra una ortofoto en tiles usando python y gdal translate

import os
from osgeo import gdal

in_path = 'C:/Users/john/Downloads/AutoMap/datos/area/'
input_filename = 'RetiroAoI.tif'

out_path = 'C:/Users/john/Downloads/AutoMap/datos/inference1000/'
output_filename = 'tile_'

dset = gdal.Open(in_path + input_filename)

print("la ortofoto", dset.GetGeoTransform(), dset.GetProjection())


width = dset.RasterXSize
height = dset.RasterYSize

tilesize = 600

print(width, 'x', height, ', tilesize', tilesize)

imagen = 0
for i in range(0, width, tilesize):
    for j in range(0, height, tilesize):
        w = min(i+tilesize, width) - i
        h = min(j+tilesize, height) - j
        gdaltranString = "gdal_translate -of GTIFF -srcwin "+str(i)+", "+str(j)+", "+str(w)+", " \
            +str(h)+" " + str(in_path) + str(input_filename) + " " + str(out_path) + str(output_filename) + str(i) + "_" + str(j) + ".TIF"
        os.system(gdaltranString)
        source = gdal.Open(str(out_path) + str(output_filename) + str(i) + "_" + str(j) + ".TIF")
        
        print("tile by tile", source.GetGeoTransform(), source.GetProjection())
        imagen = imagen + 1
print('El numero de imagenes producidas es:', imagen)
print("Ojo la particion empieza en el punto cero en la esquina superior izquierda y varia la 'y' hacia abajo")
```
