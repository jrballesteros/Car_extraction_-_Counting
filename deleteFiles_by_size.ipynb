```python
# Delete files by size

import os, os.path

dirtocheck = "C:/Users/john/Downloads/PythonCourseUN/software/datasets/combined3/"

i = 1
for root, _, files in os.walk(dirtocheck):
    for f in files:
        fullpath = os.path.join(root, f)
        if os.path.getsize(fullpath) <= 12.7 * 1024:
            os.remove(fullpath)
            i = i+1
            print(f)
print('El numero total de archivos borrados fue de:', i)
```
